# Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
class triangle:
    def __init__(self):
        self.side1,self.side2,self.side3 = [float(i) for i in input('Enter sides of triangle with space ').split()]
    def area(self):
        self.side = (self.side1+ self.side2+self.side3)/2
        self.area = (self.side*(self.side-self.side1)*(self.side-self.side2)*(self.side-self.side3))**0.5
        return self.area
triangle_1 = triangle()
print("Area of the triangle is : ",triangle_1.area())
```

# Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(lst,n):
    return list(filter(lambda i:len(i)>n , lst))
list_of_countries = ['India','London','America','China','Japan','Sweden','Brazil']
filter_long_words(list_of_countries,5)
```

# Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

```python
def map_lst_length(lst):
    return list(map(lambda i: len(i),lst))
list_of_countries = ['India','London','America','China','Japan','Sweden','Brazil']
map_lst_length(list_of_countries)
```

# Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```python
def vowel(c):
    if len(c) > 1:
        print("Please enter only one character")
    elif c in ['a','e','i','o','u']:
        return True
    else:
        return False
c = input("Enter any character :")
new_c = c.lower()
print("Is {} a vowel?: ".format(c),vowel(new_c))
```
