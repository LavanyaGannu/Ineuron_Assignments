# Importing Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets 
```

# Loading IRIS Data

```python
import pandas as pd
from sklearn.datasets import load_iris
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
iris_df['Target'] = iris_data['target']
iris_df.head()
```

# Defining X and Y variables

```python
X = iris_df.iloc[:,:].values
y = iris_df.iloc[:,4].values
```

# Scaling and Transforming X and Y

```python
from sklearn.preprocessing import LabelEncoder
lab_en_y = LabelEncoder()
y = lab_en_y.fit_transform(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

# Visualizing

```python
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig,elev=45,azim=130,auto_add_to_figure=False)
fig.add_axes(ax)
ax.scatter(X[:,0], X[:,1], X[:,2], c=y,cmap=plt.cm.coolwarm)
ax.set_title('3D Chart with transformed dimensions')
plt.show()
```
