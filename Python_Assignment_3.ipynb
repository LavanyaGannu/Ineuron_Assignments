# Write a Python Program to implement your own myreduce() function which works exactly  like Python's built-in function reduce() 

```python
from functools import reduce
def add(a,b):
    return (a+b)
lst = [5,7,9,12,15]
print('Output using reduce function ',reduce(add,lst))
def my_reduce(func,seq):
    res = seq[0]
    for i in seq[1:]:
        res = func(res,i)
    return res
print('Output using my_reduce function ',my_reduce(add,lst))
```

# Write a Python program to implement your own myfilter() function which works exactly  like Python's built-in function filter()

```python
def even_num(n):
    if n%2==0:
        return True
    else:
        return False
lst = [5,8,11,12,15,18,21,24]
print("Output using filter function: ", list(filter(even_num, lst)))
def my_filter(func,seq):
    lst1 = []
    for i in seq:
        if func(i)==True:
            lst1.append(i)
    return lst1
print('Output using my_filter function ',my_filter(even_num,lst))
```
