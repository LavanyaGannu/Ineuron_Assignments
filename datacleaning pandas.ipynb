```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'], 
                    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
                    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
                    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df['FlightNumber'] = ((df['FlightNumber'].fillna(0) + 10).astype(int))
```

```python
temp_df= df.From_To.str.split('_',expand = True)
temp_df.columns = ['From','To']
```

```python
temp_df['From'] = temp_df['From'].str.capitalize()
temp_df['To'] = temp_df['To'].str.capitalize()
```

```python
df = df.drop(columns='From_To')
df = df.join(temp_df)
df
```

```python
split_df = pd.DataFrame(df['RecentDelays'].tolist(), columns=['delay_1', 'delay_2', 'delay_3'])
df = df.drop(columns='RecentDelays')
df = df.join(split_df)
```
